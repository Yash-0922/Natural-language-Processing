This downloads the pretrained English pipeline.

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = "I'm yash Learning NLP today!!"
doc = nlp(text)
for token in doc:
  print(token.text,token.i, len(token),token.lower_,token.whitespace_)
```

It is a Doc object.
A Doc behaves a bit like a list of tokens.

```python
text = "Python 3.12 is amazing!"

doc = nlp(text)

for token in doc:
    print(
        token.text,
        token.is_alpha,
        token.is_digit,
        token.is_punct,
        token.like_num,
        token.shape_
    )
```

```python
text = "Heyy yash is 23 years old "
text2 = "boss is won ten dollars"

doc = nlp(text2)
for token in doc:
  if token.like_num:
    print(f'found -> {token.text}')
    break
```

Sentence Segmentation

spaCy can identify sentence boundaries.

```python
text = "I love Maths. Math is fascinating"
doc = nlp(text)
for i in doc.sents:
  print(i)
```

```python
reviews = [
    "Amazing product!",
    "Delivery was slow.",
    "Worth every penny."
]
```

```python
for review in reviews:
    doc = nlp(review)

    tokens = [token.text for token in doc]
    print(tokens)
```

```python
reviews = [
    "Amazing product!",
    "Delivery was slow.",
    "Worth every penny."
]

for review in reviews:
  doc = nlp(review)
  new_tokens = []
  for token in doc:
     new_tokens.append(token.text)
  print(new_tokens)

```

```python
reviews = [
    "Amazing product!",
    "Delivery was slow.",
    "Worth every penny."
]

for review in reviews:
    doc = nlp(review)

    new_tokens = []

    for token in doc:
        new_tokens.append(token.text)

    print(new_tokens)
```

1. Token Indexing

```python
text = "Me Yash patil"
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
print(doc[0])
print(doc[1])
print(doc[2])
print(doc[-2])


```

2. Slicing

```python
text = "My name is yash patil"
doc = nlp(text)
print(doc[1:4])
print(doc[-1:-1])
print(doc[:0])
```

3. Iteration

```python
text = "I'm yash patil learning NLP"
doc = nlp(text)
for i in doc:
  print(i.text)
```

```python
for i , token in enumerate(doc):
  print(i,token.text)
```

4. Character Offsets
This is one of spaCy's most useful features.

```python
text = "My name is Yash Patil"
doc = nlp(text)
for token in doc:
    print(token.text, token.idx)
```

5. Token Length

```python
for token in doc:
    print(token.text, len(token))
```

6. Whitespace

spaCy remembers spaces.

```python
text = "Python   is awesome."

doc = nlp(text)

for token in doc:
    print(
        token.text,
        repr(token.whitespace_)
    )
```

7. Token Shape

```python
doc = nlp("GPT4 Python OPENAI OpenAI")

for token in doc:
    print(token.text, token.shape_)
```

9. Lexeme

This is a concept many beginners skip.

```python
lexeme = nlp.vocab["Python"]

print(lexeme.text)
```

```python
lexeme.is_alpha
lexeme.is_stop
lexeme.shape_
```

```python
reviews = [
    "Amazing product!",
    "Delivery was slow.",
    "Worth every penny.",
    "Packaging was damaged."
]

docs = nlp.pipe(reviews)

for doc in docs:
    print("-" * 30)

    for token in doc:
        print(
            token.text,
            token.idx,
            token.is_alpha,
            token.shape_
        )
```
