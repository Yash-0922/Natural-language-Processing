**Whitespace Tokenization** splits text whenever it encounters whitespace characters.

```python
text = "Yash     Kailas"
token = text.split()
print(token)
```

```python
from nltk.tokenize import WhitespaceTokenizer
tokenizer = WhitespaceTokenizer()
text = "I love python     and  math  s"
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
text = "Python\nNLP\tAI"
token = text.split("\n")
print(token)
```

```python
text ="Python\nNLP\tAI"
tokenizer = WhitespaceTokenizer()
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
text = "Hello,\nWorld!"

print(text.split())
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

text = "Hello, #world!"

print(tokenizer.tokenize(text))
```

```python
tokenizer = RegexpTokenizer(r'\d+')

text = "Apple costs 500 rupees."

print(tokenizer.tokenize(text))
```

```python
tokenizer = RegexpTokenizer(r'[A-Za-z]+')

text = "AI2026 Python3"

print(tokenizer.tokenize(text))
```
