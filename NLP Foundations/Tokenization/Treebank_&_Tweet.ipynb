**Treebank**

It comes from the Penn Treebank, a large annotated English corpus created at the University of Pennsylvania.

That dataset contains:

Sentences
Part-of-Speech tags
Parse trees
Syntactic annotations

To keep annotation consistent, researchers defined tokenization rules. Those rules became the Treebank tokenizer.

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize("I'm Yash Patil"))
```

```python
text = "Heyy Rahul It's me Yash "
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(text))
```

TweetTokenizer


TweetTokenizer was designed for social media.
TweetTokenizer was specifically designed to handle below cases more naturally.
                     
                     
Emojis
Hashtags
Mentions
URLs
Repeated characters
Emoticons




basic diff between tree and tweet tokenizer

```python
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TweetTokenizer()
tokenizer1 = TreebankWordTokenizer()
text = "heyyy Yashh !! How're you ?"

print(f'Tweet -> {tokenizer.tokenize(text)}')
print(f'tree -> {tokenizer1.tokenize(text)}')
```

repeated characters

```python
text = "Sooooooo happy"

print(tokenizer.tokenize(text))
```

repeated characters using paramters

parameter 1 -> reduce_len

```python
tokenizer = TweetTokenizer(reduce_len=True)

text = "Sooooooo happy"

print(tokenizer.tokenize(text))
```

parameter 2-> strip_handles

```python
tokenizer = TweetTokenizer(strip_handles=True)

text = "@OpenAI released GPT."

print(tokenizer.tokenize(text))
```

Parameter 3 -> preserve_case

```python
tokenizer = TweetTokenizer(preserve_case=False)

text = "I LOVE PYTHON"

print(tokenizer.tokenize(text))
```
